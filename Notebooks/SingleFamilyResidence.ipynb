```python
import pandas as pd
import numpy as np
import csv
import os
import io
import csv
import plotly.plotly as py
import plotly.figure_factory as ff
```

```python
csv_path = os.path.join("Zip_Zhvi_SingleFamilyResidence2.csv")
sfr = pd.read_csv(csv_path)
sfr.head()
```

```python
kcmetrosfr1 = sfr.loc[sfr["Metro"] == "Kansas City"]
kcmetrosfr1
#attended_bootcamp = reduced_coders_pd.loc[reduced_coders_pd["AttendedBootcamp"] == "Yes"]
#attended_bootcamp.head()
```

```python
kcmetro = kcmetrosfr1.loc[sfr["CountyName"].isin (["Jackson", "Johnson", "Wyandotte", "Clay", "Platte"])]
kcmetro
```

```python
kckmetro = kcmetro.loc[kcmetro["State"] == "KS"]
kckmetro
```

```python
groupKS = kckmetro.groupby(["City"])
groupKS.mean().round(0)
# Create the GroupBy object based on the "Type 1" column
#pokemon_group = pokemon_type.groupby(["Type 1"])

# Calculate averages for combat stats using the .mean() method
#pokemon_group.mean()
```

```python
groupKS1 = groupKS.mean().round(0)
groupKS1
```

```python
groupKS1.reset_index(inplace=True)
groupKS1
```

```python
kcmmetro = kcmetro.loc[kcmetro["State"] == "MO"]
kcmmetro
```

```python
groupMO = kcmmetro.groupby(["City"])
groupMO.mean().round(0)
```

```python
groupMO1 = groupMO.mean().round(0)
groupMO1
```

```python
groupMO1.reset_index(inplace=True)
groupMO1.head()
```

```python
file_name = os.path.join("MO_sfr_cg.csv")
groupMO1.to_csv(file_name, index=False)
```

```python
file_name = os.path.join("KS_sfr_cg.csv")
groupKS1.to_csv(file_name, index=False)
```
